```python
tomato = pd.read_csv("https://raw.githubusercontent.com/UCD-pbio-rclub/python_problems/master/Tomato.csv")
tomato.head()
```

```python
subtomato = tomato[["trt","hyp","species"]]
print(subtomato.head())
print()
# This also works:
print(tomato.loc[:,['trt', 'hyp', 'species']].head())
```

```python
subChile = subtomato[subtomato.species=="S. chilense"]
subChile.head()
```

Base or numpy:

```python
print (subChile[subChile.trt=="H"]["hyp"].mean())

print (subChile[subChile.trt=="L"]["hyp"].mean())
```

```python
np.mean(subChile[subChile.trt=="H"]["hyp"])
np.mean(subChile[subChile.trt=="L"]["hyp"])
```

```python
## Dictionary comprehension
{each: subChile[subChile.trt==each]["hyp"].mean() for each in np.unique(subChile.trt)}
```

```python
## List comprehension
[subChile[subChile.trt==each]["hyp"].mean() for each in np.unique(subChile.trt)]
```

```python
## Set comprehension
{np.mean(subChile[subChile.trt==each]["hyp"]) for each in np.unique(subChile.trt)}
```
