```python
url = ' https://api.openbrewerydb.org/breweries?&per_page=50'
resp = requests.get(url)
resp
data = resp.json()
```

```python
## Create a dataframe using the api address above with every brewery in the database. 
# Will need to use page and per_page as parameters. per_page max is 50
brewDF = pd.DataFrame(data)
brewDF.shape
```

```python
brewDF.head()
```

```python
## Filter this data set down to only micro breweries in states that with begin and end with the same letter
estados = brewDF.state
brew_match = brewDF[[each[0].lower()==each[-1].lower() for each in estados]]
```

```python

brew_match.head()
```

```python
brew_match[(brew_match.latitude == max(brew_match.latitude)) | (brew_match.latitude == min(brew_match.latitude)) |
          (brew_match.longitude == max(brew_match.longitude)) | (brew_match.longitude == min(brew_match.longitude))]
```
